```python
employee_pay = [15000, 120000, 35000, 45000]
count = 0
sum = 0
```

```python
for index in range(0, len(employee_pay)):
    count += 1
    sum += employee_pay[index]
    
average = sum / count
print('The total payroll is: ', sum)
print('The total payroll is: ', average)
```

Soal 3

```python
from random import randint

target = randint(1,10)
chance = 1
print('Guess an integer from 1 to 10. You will have 3 chance.')
while chance <= 3:
    guess = int(input('Guess an integer'))
    if guess > target:
        print('Guess is to high')
    elif guess < target:
        print('Guess is to low')
    else:
        print('Guess is just right!')
        break
    chance += 1
```
